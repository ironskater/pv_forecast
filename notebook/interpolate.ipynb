```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
num_of_point = 50
x = np.linspace(start = 0, stop = np.pi, num = num_of_point)
y = np.sin(x)
dt_index = pd.date_range('2021-07-27 08:00', freq = '2T', periods = num_of_point)

df = pd.DataFrame({'y':y}, index = dt_index)

df.plot(figsize=(12, 8))
```

```python
df.iloc[20:49] = None

df.plot(figsize=(12, 8))
```

```python
df_missing = df.copy()

df_missing.interpolate(inplace = True)

df_missing.plot(figsize=(12, 8))
```

```python
df_missing = df.copy()

df_missing.interpolate(method = 'polynomial', order = 2, inplace = True)

df_missing.plot(figsize = (12, 8))
```
